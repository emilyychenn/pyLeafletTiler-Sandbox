```python
# my computer seems to be missing the libpng delegate for imagemagick; can only read png
# magick identify -list format --> command (command line) to show which file formats can be read
# therefore must convert image to jpg before making into tiles
tile_width = 256
tile_height = 256
image_width = 4096
image_height = 4096 # image must be square in order to generate non-stretched map tiles!!!
zoom_level = 5


while zoom_level>0:
    !mkdir $zoom_level 2> /dev/null
    
    cmd = 'convert CD31.jpg -scale {0}x{1}\! {2}/full_zoom.png'.format(
            image_width, image_height, zoom_level)
    print(cmd)
    os.system(cmd) # to run the command
    
#     cmd2 = 'convert multiply.png -scale $image_width x$image_height\! $zoomlevel/full_zoom.png'
#     print(cmd2)
#     os.system(cmd2)

    cmd3 = 'convert -crop 256X256 +repage {0}/full_zoom.png {0}/tiles_%d.png'.format(zoom_level)
    print(cmd3) 
    os.system(cmd3)
   
    zoom_level -= 1
    image_width = image_width//2
    image_height = image_height//2
  
```

```python
image_width = 4096
zoom_level = 5

while zoom_level>0:
    fname = !ls "$zoom_level"/tiles_*.png
    total_tiles = len(fname) # To get this number, look at the number of tiles generated, find the last tile number & add 1
    print(total_tiles, zoom_level) # e.g. tiles_99.png => total_tiles = 100
    tiles_per_column = image_width/tile_width

    row = 0
    column = 0
    for i in range(total_tiles):
        filename = "{0}/tiles_{1}.png".format(zoom_level, i) # current filename
        target = "{0}/map_{1}_{2}.png".format(zoom_level, column, row) # new filename

        print('cp -f {0} {1}'.format(filename, target))
        !mv $filename $target

        column = column + 1
        if column >= tiles_per_column:
            column = 0
            row = row + 1
            
    image_width = image_width//2
    zoom_level -= 1
```
